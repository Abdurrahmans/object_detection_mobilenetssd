```python
import numpy as np
import cv2
```

```python
webcom_videos_stream = cv2.VideoCapture('images/videoplayback.mp4')
while True:
    ret,current_frame = webcom_videos_stream.read()
    img_to_detect = current_frame
    img_height = img_to_detect.shape[0]
    img_width = img_to_detect.shape[1]
    resige_img_to_detect = cv2.resize(img_to_detect,(300,300))
    img_blog  = cv2.dnn.blobFromImage(resige_img_to_detect,0.007843,(300,300),127.5)

    class_labels =["background", "aeroplane", "bicycle", "bird", "boat","bottle",
                   "bus","car", "cat", "chair", "cow", "diningtable","dog", "horse",
                   "motorbike", "person", "pottedplant", "sheep","sofa", "train", "tvmonitor"]

    imagenetssd = cv2.dnn.readNetFromCaffe('mobilenet/MobileNetSSD_deploy.prototxt','mobilenet/MobileNetSSD_deploy.caffemodel')
    imagenetssd.setInput(img_blog)
    obj_detection = imagenetssd.forward()
    for index in np.arange(0, obj_detection.shape[2]):
        prediction_confidence = obj_detection[0,0,index,2]
        if prediction_confidence > 0.20:
            predicted_class_index = int(obj_detection[0,0,index,1])
            predicted_class_label = class_labels[predicted_class_index]
            bounding_box = obj_detection[0,0,index, 3:7] * np.array([img_width,img_height, img_width,img_height])
            (startX,startY,endX,endY) = bounding_box.astype("int")
            predicted_class_label = "{}: {:.2f}%".format(class_labels[predicted_class_index], prediction_confidence*100)
            print("predicted object {}: {}".format(index+1, predicted_class_label))
            cv2.rectangle(img_to_detect,(startX,startY),(endX,endY),(0,255,0),2)
            cv2.putText(img_to_detect,predicted_class_label,(startX,startY-5),cv2.FONT_HERSHEY_TRIPLEX,0.7,(0,255,0),1)


    cv2.imshow("detect output",img_to_detect)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
     

    
webcom_videos_stream.release()
cv2.destroyAllWindows()
```
